```python
data['target'] = pd.cut(data.strength, bins=[0,23,46,83], labels=[1,2,3])
```

Since we have continous values for our target columm, i have decided to split them in to three categories.
1-bad/poor
2-good
3-best

```python
x = data[data.columns[0:-2]]
y = data["target"]
```

```python
from scipy.stats import zscore
XScaled=x.apply(zscore)
XScaled.head()
```

```python
print(np.where(XScaled>3))
```

```python
data = data[(XScaled < 2.6).all(axis=1)]
```

I remove the OutLiers using this method.

```python
x = data[data.columns[0:-2]]
y = data["target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state = 1)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1)
```

```python
X_val.shape
```

```python
pipe = make_pipeline(MinMaxScaler(),(SVC()))
```

```python
pipe1=make_pipeline(MinMaxScaler(),DecisionTreeClassifier())
```

```python
pipe2=make_pipeline(MinMaxScaler(),RandomForestClassifier())
```

```python
param_dist = {"decisiontreeclassifier__max_depth": [3, None],
              "decisiontreeclassifier__max_features": sp_randint(1, 8),
              "decisiontreeclassifier__min_samples_split": sp_randint(2, 8),
              "decisiontreeclassifier__min_samples_leaf": sp_randint(1, 8),
              
              "decisiontreeclassifier__criterion": ["gini", "entropy"]}
```

```python
param_distr = {"randomforestclassifier__max_depth": [3, None],
              "randomforestclassifier__max_features": sp_randint(1, 8),
              "randomforestclassifier__min_samples_split": sp_randint(2, 8),
              "randomforestclassifier__min_samples_leaf": sp_randint(1, 8),
              
              "randomforestclassifier__criterion": ["gini", "entropy"]}
```

```python
param = {
              "svc__gamma": [0.01,0.1,1,10],
              "svc__C": [1, 10 ,100],
             
              }
```

```python
paramr = {
              "svc__gamma": sp_randint(0.01 , 10),
              "svc__C": sp_randint(1 , 100)
              }
```

# SVM 

```python
grid = GridSearchCV(pipe, param_grid=param,cv=10)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.cv_results_['mean_test_score']
```

```python
grid.best_estimator_
```

```python
grid.score(X_test,y_test)
```

Accuracy of Support vector model using GridSearch

```python
grid.score(X_val,y_val)
```

```python
randomCV = RandomizedSearchCV(pipe, param_distributions=paramr, n_iter=10,cv=5) 
```

```python
randomCV.fit(X_train, y_train)

 
print(randomCV.best_params_)
```

```python
randomCV.score(X_test, y_test)
```

Accuracy of Support Vector Model using RandomSearchCV

```python
randomCV.score(X_val, y_val)
```
